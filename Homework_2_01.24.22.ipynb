```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
start = -1 * math.pi #starting point, because np and range doesn't work with floats
x = [start] #x array

#create the array here
while start < math.pi:
    start += 0.1
    x.append(start)
    
```

```python
print(x)
```

```python
#define the function that matches the one given in class
def func(count, x):
    """
    input: number of sin waves added together (count), x-values (x)
    """
    out = [] #save the y values here
    for v in x:
        sm = (math.pi/2) #sum
        n = 1 #start with this number of sine waves
        add = 0 #the sum of all the sine waves
        while n <= count:
            add += np.sin(n*v)/n #sin(nx)/n
            n += 2 #move up by 2
        add = add * 2 #multiply the sum of the sines by 2
        out.append(sm + add) #add this value to the output
        
    return out
```

```python
y1 = func(11, x) #call the function
```

```python
print(y1)
```

```python
#graph the function, n=11
plt.plot(x, y1)
plt.show()
```

```python
#call and graph the function, n=21
y2 = func(21, x)
plt.plot(x, y2)
plt.show()
```

```python
#call and graph the function, n=31
y3 = func(31, x)
plt.plot(x, y3)
plt.show()
```

```python
#This will plot them all on one plot
plt.plot(x, y1, "g")
plt.plot(x, y2, "r")
plt.plot(x, y3, "b")
plt.legend(["11", "21", "31"]) #create a label
plt.show()
```
