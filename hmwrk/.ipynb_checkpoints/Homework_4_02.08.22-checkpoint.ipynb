## Assignment
### -- Forward Difference Scheme --

```python
#This is the forward differecne methods
def forwardMeth(f, dx, x=math.pi/4):
    return (f(x+dx)-f(x))/dx #return the difference between the current and next step in the function
```

```python
ana = lambda x: -1 * np.sin(x) #This is the analytical derivative
```

```python
err = list() #This will be the error list log10(numerical - analytical)
dx = list() #this will hold the log10(dx)
for i in range(1, 17, 1):
    fw = forwardMeth(f, 10**(-1*i))
    a = ana(math.pi/4)
    err.append(np.log10(np.abs(fw-a))) #This will store the np.log(numerial - analytical) error
    dx.append(np.log10(10**(-1*i))) #this is the np.log(dx)
```

```python
plt.plot(dx, err)
plt.ylabel("log10(numerical - analytical)")
plt.xlabel("log10(dx)")
plt.show()
```

### --Analysis--

Since the error will look roughly like ~$dx^n$ where n is the magnitude of the step value. The best possible step value for the forward difference method is where n=8 or $dx^8$. This is because the error for the forward difference method is $O(h)$, so it is linear. After n=8 though, the difference between numbers becomes two small, and the substraction of the values will show a greater error.

### --Central Difference Scheme--

```python
#This will define the central difference scheme 
def centralMeth(f, dx, x=math.pi/4):
    return (f(x+(dx/2))-f(x-(dx/2)))/dx
```

```python
#Redefine it here
ana = lambda x: -1 * np.sin(x) #This is the analytical derivative
```

```python
err = list() #This will be the error list log10(numerical - analytical)
dx = list() #this will hold the log10(dx)
for i in range(1, 17, 1):
    fw = centralMeth(f, 10**(-1*i))
    a = ana(math.pi/4)
    err.append(np.log10(np.abs(fw-a))) #This will store the np.log(numerial - analytical) error
    dx.append(np.log10(10**(-1*i))) #this is the np.log(dx)
```

```python
plt.plot(dx, err)
plt.ylabel("log10(numerical - analytical)")
plt.xlabel("log10(dx)")
plt.show()
```

#

### --Analysis--

Well, this scheme is better then the foward-difference method because the central difference doesn't subtract two numbers that are very close to one another. It spreads out the difference in the numerator between two different numbers, which makes it easier for the machine to calculate the difference without a truncation error. This makes the answer more accurate to the analytical method. This means that the error is of a smaller magnitude with smaller dx intervals, which makes it easier to process. It also has a error that follows $O(h^2)$, which is much better then the forward difference error, because this means the error gets smaller faster then the previous method.

The error for the central difference is given by the following sequence

\begin{equation}
\frac{h^2}{4*3!}*y'''(t)+\frac{(h^4}{2^4*5!}*y^(5)(t)....

which if we plug in cos(x) in to the error, we'll consistantly get for each derivative a sin(0).
