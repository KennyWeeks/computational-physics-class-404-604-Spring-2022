Any mass can lead to a gravitational potential in 3-D (x-y-z) space
\begin{equation}
\nabla^2\Phi=4\pi G\rho
\end{equation}
    

We will set G=1 for this problem.  
Please solve the potential of a line mass 
(i.e. one zone at x=50, y=50 with nonzero density and all others with zero density) 
using SOR. Since the problem has symmetry in the Z direction:
\begin{equation}
\nabla^2\Phi=\frac{\partial^2\Phi}{\partial^2 x}+\frac{\partial^2\Phi}{\partial^2y}
\end{equation}


The domain is X from 0 to 100, Y from 0 to 100. $\Delta$x= $\Delta$y=1. (assume $\Phi=0$
at the boundary and $\rho_{50,50}$ is always 1)

Please verify that the potential has radial dependence close to 2*G*$\mu$*ln(r)+constant. 
Where $\mu$ is its mass per unit length and r is the distance to the line mass. 



### SOR Method

```python
import matplotlib.pylab as p;
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy as np;
print("Initializing")
```

```python
# This is just some constant declarations
omega = 1.2 #the omega constant in the sor method
Nmax = 100 #the maximum in both the x and y direction
Niter = 3000 #this is the the number of iterations that the sor method will run through
GP = zeros((Nmax, Nmax), float) #the array of the gravitational potentials generated by the line mass

GP[50, 50] = 4 * np.pi * 1 * 1 #this is the gravitational potential at the center where the density is 1 and the G=1

for iter in range(Niter):
    if iter%10 == 0: print(iter) #This is just to keep track of my iterations
    maxerr = 0.
    for i in range(1, Nmax-1):
        for j in range(1, Nmax-1):
            du = 1/4*(GP[i+1, j] + GP[i-1, j] + GP[i, j+1] + GP[i, j-1]) - GP[i, j]
            if maxerr < abs(du):
                maxerr = abs(du)
            GP[i, j] = GP[i, j] + omega * du
    print(maxerr) 
    if(maxerr < 1.e-7):
        break
```

```python
#begin graphing stuff
x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X, Y = p.meshgrid(x,y)                 

def functz(GP):                                         # V(x, y) 
    z = GP[X,Y]                        
    return z

Z = functz(GP)                          
fig = p.figure()                                      # Create figure
ax = Axes3D(fig)                                      # Plot axes
ax.plot_wireframe(X, Y, Z, color = 'r')               # Red wireframe
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show() 
```

### Proof of topology

This is just proving that the above graph of the gravitational potential is close to the following equation

2*G*$\mu$*ln(r)+constant

Where $\mu$ is the mass per s length. From the previous equation, we know that the density at (50, 50) is 1, so that mean that the mass is the same magnitude as the volume. So if the line was projected into a 3-d plane, we know that the magnitude of the mass would equal $mass=\pi*{r^2}*h$ and that $\mu=\frac{mass}{h}=\pi*{r^2}$ where r is some constant. We can set that to something really small, considering that this is a point in 2-d space.

```python
radius = zeros((Nmax, Nmax), float)

x = 0
while x < 50:
    y = 0
    while y < 50:
        rad = (x**2 + y**2)**(1/2)
        radius[x, y] = rad
        y += 1
    x+=1

x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X, Y = p.meshgrid(x,y)

def functz(radius): 
    z = zeros((Nmax, Nmax), float)# V(x, y) 
    for i in range(len(radius)):
        for j in range(len(radius[i])):
            z[i, j] = 2 * 1 * np.pi * (1.e-4)**2 * np.log(radius[i, j])-1
    return z

Z = functz(radius)                          
fig = p.figure()                                      # Create figure
ax = Axes3D(fig)                                      # Plot axes
ax.plot_wireframe(X, Y, Z, color = 'g')               # Red wireframe
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show() 
```
