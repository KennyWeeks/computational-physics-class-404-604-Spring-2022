### Burger's Equation

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Model Parameters
xmin = -10.0   # left boundary
xmax = +10.0   # right boundary
Nx = 101      # number of grid points (including boundary)
tend = 10.0    # end time
dtout = 1.0   # time interval for outputs

# Set up the grid.
x = np.linspace(xmin, xmax, Nx)
dx = (xmax - xmin) / (Nx - 1)
dt = 0.8 * dx
U = np.zeros(Nx,)

# Give the initial profile.
t = 0.0
#U = 0.2 + 0.8 * np.exp(-0.5 * x**2)

#If UFx defines the velocity here, then we know we'll never have a negative velocity

UFx = lambda x: 0.2 + 0.8*np.exp(-0.5*x**2) #this is the lambda function for this program
UL = [UFx(d) for d in x] #this is the initial U values, also this will be the list used in the upstream method 
```

```python
fig = plt.figure(figsize=(16,10))
tout = t + dtout
plt.plot(x, UL)
tp = [t]
Up = np.copy(UL)
while t < tend:
    #The upstream method begins here
    Uold = np.copy(UL)
    
    #The velocity will never be less then zero, it will only change exponentially
    for ix in range(1, Nx - 1):
        UL[ix] = Uold[ix] - (Uold[ix]*dt)*(Uold[ix] - Uold[ix-1])/dx
    
    t += dt
    
    #saving the data for each step
    if t >= tout:
        tp.append(t)
        plt.plot(x, UL, label=r"$t = {:.3G}$".format(t))
        Up = np.vstack((Up, UL))
        print("t = ", t)
        tout += dtout
        
```

```python
# Create 3D-view of the solution.
t, x = np.meshgrid(tp, x)
fig3D = plt.figure(figsize=(16,10))
ax3D = Axes3D(fig3D)
ax3D.plot_wireframe(t, x, Up.transpose(), color="red")
ax3D.set_xlabel("Time $t$")
ax3D.set_ylabel("Position $x$")
ax3D.set_zlabel(r"$U(t,x)$")
```

### Analysis 

The solution develops a non-symetric profile over time because with each advection of the model forward, it's velocity is the value of the previous time step at the position multiplied by a differential step value. Because of this, as the model move forward, the model's shape will change. This stems from a time-dependent velocity, or a non-constant velocity.
